```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

%matplotlib inline 
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('../data/clean_df.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.index = df['date'] 
df = df.resample('M').sum()
```

```python
df
```

# Seasonality

```python
decomp = seasonal_decompose(df['civilian_casualties'])
```

```python
decomp.plot();
plt.tight_layout()
plt.savefig('../images/civilian_casualties_seasonal_decomp.png')
```

```python
decomp.seasonal.plot()
plt.title('Civilian Casualties Seasonal Decomposition')
plt.xlabel('Year')
plt.ylabel('Civilian Casualties');
plt.tight_layout()
plt.savefig('../images/Civ_casualties_seasonal_decomp_large.png')
```

```python
plot_pacf(df['civilian_casualties'], lags=12, alpha=.05);
```

```python
plot_acf(df['civilian_casualties'], lags=12, alpha=.05);
```

# Checking Stationarity

```python
df['first_diff'] = df['civilian_casualties'].diff()
```

```python
plt.plot(df['first_diff'])
plt.title('First Differenced Civilian Casualty Data');
```

# Arima Modeling

```python
y_train, y_test = train_test_split(df['civilian_casualties'].diff(), random_state=6, shuffle=False, test_size=.1)
```

```python
model = ARIMA(endog=y_train, order=(0, 1, 2)).fit()
predictions = model.predict(start=y_test.index[0], end=y_test.index[-1])
predictions
```

```python
# Plotting the Model
plt.figure(figsize=(10,6))
plt.plot(y_train.index, y_train, color = 'blue') #training data
plt.plot(y_test.index, y_test, color = 'orange') # testing data
plt.plot(y_test.index, predictions, color = 'green') # predictions based on testing data
plt.title('Civilian Casualties over time')
plt.xlabel('Time');
```
